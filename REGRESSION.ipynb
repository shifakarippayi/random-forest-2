**Classification and regression:** Manchine learning takes can be divided into classified regression. in classification we use catagorical labels, (yes or no) but in refression problems our prediction is countinuous numerical data.

**Linear Regression**

Linear Regression is a stastical
regression method which is used for predictive analysis

* it is one of the very simple and esay algorithms which is works on regression and shows the relationship b/w the continuous variables.
* Linear regression shows the linear relationship b/w the independed varaible(x axis), and the the dependent variable(y axis) , hence called linear regression.

**Simple Linear regession Multiple linear Regression**

IF there is only input variable (x) ,then such linear regression is called simple linear regression. is more one input variable. then such a linear regression is called multiple linera regression.

**Mathemetical Equation**

# Y= aX=b

* Y= dependent varaiable(target)
* X= independent varaible(predicted)

* a and b are linear coefficients


$
a = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$

$
b = \bar{y} - a\bar{x}
$



**Simple Linear Regession Problem**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
house_size= np.array([800,1000,1200,1500,1800])
house_prize=np.array([100,180,230,250,340,])
# feature into column wise
X=house_size.reshape(-1,1)
print(X)

```

```python
# model
model=LinearRegression()
```

```python
model.fit(X,house_prize) # tranind model(finding best fit line)
```

```python
X_new =np.array([2000]).reshape(-1,1)
```

```python
y_pred= model.predict(X_new)
```

```python
y_pred
```

```python
print(f'the house prize for 2000 house size is :{y_pred}')
```

```python
X2= np.array([2500]).reshape(-1,1)
y_pred= model.predict(X2)
y_pred
```

```python
print(f'the house prize for 2500 house size is:{y_pred}')
```

```python
X_new= np.array([700,1000,800,1200,1500,1800,2000,2500,2700]).reshape(-1,1)
predicted_values= model.predict(X_new)
predicted_values
```

```python
#regression scatter plot
import matplotlib.pyplot as plt
plt.scatter(house_size,house_prize,color='b',label='data points')
plt.plot(X_new,predicted_values,color='r',label='regression line')
plt.xlabel('house size')
plt.ylabel('house prize')
plt.legend
plt.show()
```
