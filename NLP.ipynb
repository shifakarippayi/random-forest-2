```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

sentence="at Eight o clock on Tuesday Morning, Hello How are you!!"

stop_words=set(stopwords.words('english'))
tokens=word_tokenize(sentence)
tokens
words=[w for w in tokens if not w.lower() in stop_words]
words
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
sentence= 'at Eight o Clock on Monday , Hello How Are You'
stopwordss=set(stopwords.words('english'))
tokens=word_tokenize(sentence)
tokens
words=[ W. lower() for W in tokens if not W in stopwordss]
words
```

```python
from nltk.util import ngrams
sentence= 'at Eight o Clock on Monday , Hello How Are You'
Ngrams=ngrams(nltk.word_tokenize(sentence),n=3)
for grams in Ngrams:
  print(grams)
```

```python
from nltk.stem import SnowballStemmer
sb=SnowballStemmer('english')
words=['eight', 'clock','programs','programming','programmer','program','monday','morning',',','hello']
for w in words:
  print(w, " , ",sb.stem(w))
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
words=['eight', 'clock','programs','programming','monday','morning',',','hello']
print('Words, Porter Stem')
for w in words:
  print(w, " , ",ps.stem(w))
```

**Lemmatization**

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
print('rocks',lem.lemmatize('rocks'))
print('corpora',lem.lemmatize('corpora'))
print('better',lem.lemmatize('better'))
print('cats',lem.lemmatize('cats'))
```

**part of speech tagging**: in nlp part of speech tagging is the process of assinging grammatical categories(like noun, verb, adjective,etc)

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
print('rocks as noun:',lem.lemmatize('rocks'))
print('thought as noun:',lem.lemmatize('thought'))
print('corpora as noun:',lem.lemmatize('corpora',pos='n'))
print('runing as verb:',lem.lemmatize('runing',pos='v'))
print('played as verb:',lem.lemmatize('played',pos='v'))
```

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()

print('rocks as noun:',lem.lemmatize('rocks',pos='n'))
print('thought as verb:',lem.lemmatize('thought',pos='v'))
print('play as adverb:',lem.lemmatize('play',pos='r'))
print('better as adjective:',lem.lemmatize('better',pos='a'))
```
